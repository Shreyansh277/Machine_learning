**outlier detection**

```python
def detect_outliers_iqr(dataframe):
    """
    Detects outliers in each feature of the DataFrame using the Interquartile Range (IQR) method.

    Parameters:
        dataframe (pd.DataFrame): The input DataFrame from which to detect outliers.

    Returns:
        dict: A dictionary where keys are feature names and values are lists of indices of detected outliers.
    """
    outliers_dict = {}  # Initialize a dictionary to store outlier indices for each feature

    # Iterate through each feature in the DataFrame (skipping the target column)
    for feature in dataframe.columns[1:]:
        # Calculate the first and third quartiles
        Q1 = dataframe[feature].quantile(0.25)
        Q3 = dataframe[feature].quantile(0.75)
        IQR = Q3 - Q1  # Compute the Interquartile Range (IQR)

        # Determine the lower and upper bounds for outlier detection
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers based on the defined bounds
        outliers = dataframe[(dataframe[feature] < lower_bound) | (dataframe[feature] > upper_bound)]

        # Store the indices of the outliers in the dictionary
        outliers_dict[feature] = outliers.index.tolist()

    return outliers_dict

# Detect outliers in the dataset
detected_outliers = detect_outliers_iqr(df)

# Print the number of detected outliers for each feature
print("Detected outliers in the following features:")
for feature, indices in detected_outliers.items():
    print(f"Feature {feature}: {len(indices)} outliers")
```

# SVM with Polynomial, RBF, and Custom Kernels

**Polynomial Kernel**

```python
import time
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, roc_auc_score

# Define the degrees for the polynomial kernel
degrees = [2, 3, 4]
results_poly = {}

# Loop through each degree to train the SVM with a polynomial kernel
for degree in degrees:
    # Initialize the SVM model with polynomial kernel
    svm_poly = SVC(kernel='poly', degree=degree, random_state=10)

    # Measure cross-validated accuracy
    start_cv = time.time()  # Start timing for cross-validation
    scores_poly = cross_val_score(svm_poly, features_selected, target, cv=5)
    end_cv = time.time()  # End timing for cross-validation

    # Store the mean accuracy for the current degree
    results_poly[degree] = np.mean(scores_poly)
    print(f"Degree {degree} - Cross-validated accuracy: {results_poly[degree]:.4f}")
    print(f"Cross-validation time for degree {degree}: {end_cv - start_cv:.4f} seconds")

# Fit the model for the last degree (highest degree from the loop) and evaluate
svm_poly.fit(features_selected, target)
predictions = svm_poly.predict(features_selected)

# Output classification report and calculate AUC
print(classification_report(target, predictions))
roc_auc = roc_auc_score(target, svm_poly.decision_function(features_selected))
print("AUC:", roc_auc)

# Summary of time complexities
print("\nTime Complexity Analysis:")
print("Training (Polynomial SVM): O(n^2 * m + n^3)")  # n: number of samples, m: number of features
print("Prediction (Polynomial SVM): O(n * m)")  # n: number of samples, m: number of features
```

```python
import time
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, roc_auc_score

# Tune RBF kernel with different gamma values
gammas = [0.1, 1, 10]  # List of gamma values to experiment with
results_rbf = {}  # Dictionary to store cross-validated accuracy for each gamma

# Loop through each gamma value for the SVM model
for gamma in gammas:
    # Initialize the SVM model with RBF kernel and the current gamma value
    svm_rbf = SVC(kernel='rbf', gamma=gamma, random_state=10)

    # Measure cross-validated accuracy
    start_cv = time.time()  # Start timing for cross-validation
    scores_rbf = cross_val_score(svm_rbf, features_selected, target, cv=5)  # Perform 5-fold cross-validation
    end_cv = time.time()  # End timing for cross-validation

    # Store the mean accuracy score for the current gamma
    results_rbf[gamma] = np.mean(scores_rbf)
    print(f"Gamma {gamma} - Cross-validated accuracy: {results_rbf[gamma]:.4f}")
    print(f"Cross-validation time for gamma {gamma}: {end_cv - start_cv:.4f} seconds")

# Fit the model using the last gamma value and evaluate its performance
svm_rbf.fit(features_selected, target)
predictions = svm_rbf.predict(features_selected)  # Make predictions on the training data

# Output the classification report and AUC
print(classification_report(target, predictions))
roc_auc = roc_auc_score(target, svm_rbf.decision_function(features_selected))  # Calculate AUC
print("AUC:", roc_auc)

# Summary of time complexities
print("\nTime Complexity Analysis:")
print("Training (RBF SVM): O(n^2 * m + n^3)")  # Training complexity
print("Prediction (RBF SVM): O(n * m)")  # Prediction complexity
```

**Custom Kernel**

```python
import numpy as np
import time
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.metrics.pairwise import rbf_kernel

# Define a custom kernel that is a hybrid of RBF and linear
def custom_kernel(X, Y):
    """
    Custom kernel function combining RBF and linear kernels.

    Parameters:
    X: First input array.
    Y: Second input array.

    Returns:
    A matrix containing the combined kernel values.
    """
    return rbf_kernel(X, Y, gamma=0.1) + X @ Y.T  # Hybrid of RBF and Linear

# Initialize SVM with the custom kernel
svm_custom = SVC(kernel=custom_kernel, random_state=10)

# Measure cross-validated accuracy
start_cv = time.time()  # Start timing for cross-validation
scores_custom = cross_val_score(svm_custom, features_selected, target, cv=5)  # Perform 5-fold cross-validation
end_cv = time.time()  # End timing for cross-validation

# Output the mean cross-validated accuracy
print("Custom Kernel Results - Cross-validated accuracy:", np.mean(scores_custom))
print(f"Cross-validation time for custom kernel: {end_cv - start_cv:.4f} seconds")

# Fit the model for evaluation
svm_custom.fit(features_selected, target)  # Fit the SVM model with the selected features
predictions = svm_custom.predict(features_selected)  # Make predictions on the training data

# Output the classification report and AUC
print(classification_report(target, predictions))  # Detailed performance report
roc_auc = roc_auc_score(target, svm_custom.decision_function(features_selected))  # Calculate AUC
print("AUC:", roc_auc)

# Summary of time complexities
print("\nTime Complexity Analysis:")
print("Training (Custom SVM): O(n^2 * m + n^3)")  # Training complexity
print("Prediction (Custom SVM): O(n * m + n^2)")  # Prediction complexity
```

**Hyperparameter Tuning**

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
import time

# Define parameter grid for polynomial kernel SVM
param_grid_poly = {
    'degree': [2, 3, 4],  # Different degrees of the polynomial kernel
    'C': [0.1, 1, 10]     # Regularization parameter
}

# Initialize Grid Search for Polynomial Kernel
grid_poly = GridSearchCV(SVC(kernel='poly'), param_grid_poly, cv=5)

# Timing the grid search for polynomial kernel
start_poly = time.time()  # Start timing
grid_poly.fit(features_selected, target)  # Fit the model to find the best parameters
end_poly = time.time()  # End timing

# Output best parameters and time taken for polynomial kernel
print("Best parameters for Polynomial Kernel:", grid_poly.best_params_)
print(f"Grid Search time for Polynomial Kernel: {end_poly - start_poly:.4f} seconds")

# Define parameter grid for RBF kernel SVM
param_grid_rbf = {
    'gamma': [0.1, 1, 10],  # Different gamma values for the RBF kernel
    'C': [0.1, 1, 10]       # Regularization parameter
}

# Initialize Grid Search for RBF Kernel
grid_rbf = GridSearchCV(SVC(kernel='rbf'), param_grid_rbf, cv=5)

# Timing the grid search for RBF kernel
start_rbf = time.time()  # Start timing
grid_rbf.fit(features_selected, target)  # Fit the model to find the best parameters
end_rbf = time.time()  # End timing

# Output best parameters and time taken for RBF kernel
print("Best parameters for RBF Kernel:", grid_rbf.best_params_)
print(f"Grid Search time for RBF Kernel: {end_rbf - start_rbf:.4f} seconds")

# Summary of time complexities for both grid searches
print("\nTime Complexity Analysis:")
print("Grid Search (Polynomial Kernel): O(k * p * (n^2 * m + n^3))")  # k: number of parameter combinations, p: number of folds
print("Grid Search (RBF Kernel): O(k * p * (n^2 * m + n^3))")  # Same complexity for both kernels
```
